# I. Bunbo

##### a. Ingredient

```python
import requests
from bs4 import BeautifulSoup as bs

modern_url = 'https://bunbobae.com/category/recipes/vietnamese-recipes/modern-vietnamese/' 
response = requests.get(modern_url)

soup = bs(response.text, 'html.parser')

recipe_index_html = soup.find_all('div', class_ = 'feast-recipe-index')

modern_recipes_html = recipe_index_html[0]#có 2 class feast-recipe-index tìm được nhưng chỉ dùng cái đầu, cái sau là gợi ý món khác

modern_recipes_html_lst = modern_recipes_html.find_all("li", class_ = 'listing-item')

modern_recipes_link_list = []
for li in modern_recipes_html_lst:
    modern_recipes_link_list.append(li.find('a').get('href'))
```

```python
potato_link = modern_recipes_link_list[0]
```

```python
potato_link = modern_recipes_link_list[4]
print(potato_link)
print("-----")
response = requests.get(potato_link)
soup = bs(response.text, 'html.parser')
wp_blocks = soup.find_all('h2', class_ = 'wp-block-heading')

for wp_block in wp_blocks:
    if ('Ingredients' in wp_block.text):
        test = wp_block.find_next_sibling('ul')
        # print(test)
        ingredients = test.find_all('li')
        for ingredient in ingredients:
            print(ingredient.text)
```

# II. Hungry Huy

### 1. Get list

```python
main_dishes_page = 'https://www.hungryhuy.com/vietnamese/main-dishes-vietnamese/'
```

```python
response = requests.get(main_dishes_page)
soup = bs(response.text, 'html.parser')

dishes_html = soup.find_all('section', class_ = 'posts-grid')
print(len(dishes_html))
```

```python
dishes_img_html = dishes_html[0].find_all('div', class_ = 'entry-img')
len(dishes_img_html)
```

```python
dish_list = []
for img in dishes_img_html:
    dish_list.append(img.find('a').get('href'))
dish_list
```

### 2. Get content from link

##### a. Ingredient

```python
content_html = requests.get(dish_list[11])
soup = bs(content_html.text, 'html.parser')
```

```python
cook_instructions_html = soup.find_all('div', class_ = 'wprm-recipe-template-cutout-hungry-huy-container')
print(len(cook_instructions_html))
```

```python
ingredients_html = cook_instructions_html[0].find_all('li', class_ = 'wprm-recipe-ingredient')
print(len(ingredients_html))
```

```python
ingredients_html[2].text[1:].strip()
```

##### b. Instruction

```python
instruction_html = cook_instructions_html[0].find_all('li', class_ = 'wprm-recipe-instruction')
print(len(instruction_html))
```

# III. Recipetineats

##### a. Ingredients

```python
response = requests.get('https://www.recipetineats.com/category/vietnamese-recipes/')
soup = bs(response.text, 'html.parser')
```

```python
article_link_list = []
articles_html = soup.find_all('article', class_ = 'category-vietnamese-recipes')
for article in articles_html:
    print(article.text)
    link = article.find('a').get('href')
    print(link)
    article_link_list.append(link)
```

```python
first_article = requests.get(article_link_list[0])
first_article_soup = bs(first_article.text, 'html.parser')
```

```python
content_html = first_article_soup.find_all('div', class_ = 'wprm-entry-content')
print(len(content_html))
```

```python
ingredients_html = content_html[0].find_all('li', class_ = 'wprm-recipe-ingredient')
for ingredient in ingredients_html:
    print(ingredient.text[1:].strip())
```

##### b. Instruction

```python
instruction_html = content_html[0].find_all('li', class_ = 'wprm-recipe-instruction')
for instruction in instruction_html:
    print(instruction.text.strip())
```
